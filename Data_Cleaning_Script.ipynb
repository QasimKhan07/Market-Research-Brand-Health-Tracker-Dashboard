```
from google.colab import files
import pandas as pd

# Upload file
uploaded = files.upload()
excel_file_path = list(uploaded.keys())[0]

# Read Excel with two header rows
df = pd.read_excel(
    excel_file_path,
    sheet_name='Closed_Answers_1',
    header=[0, 1]
)

new_columns = []

for col_lvl0, col_lvl1 in df.columns:

    # ðŸ”¹ SPONT handled separately (preserve base + numbered)
    if col_lvl0 == "AU2: SPONT":
        suffix = col_lvl1.split("_")[-1]

        # keep first column as base SPONT
        if suffix == "0":
            new_columns.append("AU2: SPONT")
        else:
            new_columns.append(f"AU2: SPONT.{suffix}")

    # ðŸ”¹ BIS handled separately
    elif isinstance(col_lvl0, str) and col_lvl0.startswith("BIS_"):
        base = col_lvl0.split(":")[0].strip()   # BIS_1, BIS_2, ...
        suffix = col_lvl1.split("_")[-1]
        new_columns.append(f"{base}.{suffix}")
    elif col_lvl0 == "AD_NAME:":
        new_columns.append(col_lvl1)
    # ðŸ”¹ Everything else untouched
    else:
        new_columns.append(col_lvl0)
# ðŸ”¹ Handle AD_NAME columns correctly

df.columns = new_columns


df.head()
```
